```python
import pandas as pd 
```

```python
tasks = ['assistments', 'nhanes_lead', 'college_scorecard', 
         'brfss_diabetes', 'acsfoodstamps','diabetes_readmission',
         'brfss_blood_pressure', 'mimic_extract_los_3', 
         'mimic_extract_mort_hosp','acsincome','acspubcov', 
         'physionet', 'acsunemployment'
        ]

final_ds = []
```

```python
for task in tasks: 
    print(task)
    df = pd.read_csv(f'experiments_causal/{task}_eval.csv')
    df.drop(['Unnamed: 0', 'number'], axis=1, inplace=True)
    df['task'] = task
    df = df[(df.features=='all')]
    df = df[df.ood_test == df.ood_test.max()].reset_index(drop=True)
    if df.shape[0] > 1: 
        if 'xgb' in list(df.model): 
            df =  df[df.model == 'xgb'].reset_index(drop=True)
        else: 
            print(df.model.value_counts())
            raise ValueError('Not sure what to pick')
            
    final_ds.append(df)
```

```python
final_ds = pd.concat(final_ds, ignore_index=True)
```
